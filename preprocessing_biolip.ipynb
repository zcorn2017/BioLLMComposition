```python
#!/usr/bin/env python3
"""
Filter BioLiP files to keep only protein-DNA interactions.
Lines where column 5 (ligand type) is "dna" are kept.
"""

import sys
from pathlib import Path

def filter_biolip_file(input_file, output_file):
    """
    Filter BioLiP file to keep only lines where column 5 (ligand type) is "dna".
    
    Args:
        input_file: Path to input BioLiP file
        output_file: Path to output filtered file
    """
    input_path = Path(input_file)
    output_path = Path(output_file)
    
    if not input_path.exists():
        print(f"Error: Input file {input_file} does not exist")
        return False
    
    print(f"Processing {input_file}...")
    
    kept_count = 0
    total_count = 0
    
    with open(input_path, 'r') as infile, open(output_path, 'w') as outfile:
        for line in infile:
            total_count += 1
            if total_count % 100000 == 0:
                print(f"  Processed {total_count:,} lines, kept {kept_count:,} DNA interactions...")
            
            # Split by tab
            fields = line.rstrip('\n').split('\t')
            
            # Check if column 5 (index 4) is "dna"
            if len(fields) > 4 and fields[4].strip().lower() == 'dna':
                outfile.write(line)
                kept_count += 1
    
    print(f"  Completed: {total_count:,} total lines, {kept_count:,} DNA interactions kept")
    print(f"  Output written to {output_file}")
    return True


def main():
    # File paths
    biolip_nr_input = "/home/zcorn/Projects/proteinDNA_data/working/biolip2/BioLiP_nr.txt"
    biolip_nr_output = "/home/zcorn/Projects/proteinDNA_data/working/biolip2/BioLiP_nr_filtered.txt"
    
    biolip_input = "/home/zcorn/Projects/proteinDNA_data/working/biolip2/BioLiP.txt"
    biolip_output = "/home/zcorn/Projects/proteinDNA_data/working/biolip2/BioLiP_filtered.txt"

    
    print("=" * 60)
    print("Filtering BioLiP files to keep only protein-DNA interactions")
    print("=" * 60)
    
    # Filter BioLiP_nr.txt
    success1 = filter_biolip_file(biolip_nr_input, biolip_nr_output)
    
    if success1:

        print(f"✓ Successfully filtered BioLiP_nr.txt")
    
    print()
    
    # Filter BioLiP.txt
    success2 = filter_biolip_file(biolip_input, biolip_output)
    
    if success2:
    
        print(f"✓ Successfully filtered BioLiP.txt")
    
    print()
    print("=" * 60)
    if success1 and success2:
        print("All files filtered successfully!")
    else:
        print("Some files failed to process. Check errors above.")
        sys.exit(1)


if __name__ == "__main__":
    main()

```

```python
def add_nt_to_biolip(df_biolip_nr, dna_nr_fasta_records):
    """
    Given a DataFrame df_biolip_nr and a list of dna_nr_fasta_records,
    creates an 'nt' column in the DataFrame containing the mapped DNA sequence 
    (by pdb_id and ligand_chain). Reports the unmapped entries.
    """
    # Create an index for looking up DNA sequences based on pdb_id and chain
    dna_seq_index = {}
    for record in dna_nr_fasta_records:
        header_parts = record.id.split("_")[:3]
        if len(header_parts) >= 2:
            # e.g., '>10mh_dna_B' or '10mh_dna_B'
            # Parse id parts: e.g. '10mh_dna_B'
            pdbid = header_parts[0].lower()
            chain = header_parts[2]
            dna_seq_index[(pdbid, chain)] = str(record.seq)

    # Map function for new column
    def get_nt_seq(row):
        key = (row['pdb_id'].lower(), row['ligand_chain'].strip())
        return dna_seq_index.get(key, None)

    df_biolip_nr['nt'] = df_biolip_nr.apply(get_nt_seq, axis=1)

    # Report unmapped entries
    unmapped = df_biolip_nr[df_biolip_nr['nt'].isnull()]
    if not unmapped.empty:
        print("Entries in df_biolip_nr that could not be mapped to a DNA sequence (by pdb_id and ligand_chain):")
        print(unmapped[['pdb_id', 'ligand_chain']])
    else:
        print("All entries were mapped to nt sequence.")
    return df_biolip_nr

# Example usage:
# df_biolip_nr = map_biolipnr_to_nt(df_biolip_nr, dna_nr_fasta_records)
```
